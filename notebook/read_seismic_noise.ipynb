```python
import kontrol
import numpy as np
import matplotlib.pyplot as plt


tf1 = kontrol.load_transfer_function("tf_xg_20230722.pkl")
tf2 = kontrol.load_transfer_function("tf_xg_20240107.pkl")
tf3 = kontrol.load_transfer_function("tf_xg_20240222.pkl")

f = np.logspace(-3, 3, 1024)
plt.loglog(f, abs(tf1(1j*2*np.pi*f)))
plt.loglog(f, abs(tf2(1j*2*np.pi*f)))
plt.loglog(f, abs(tf3(1j*2*np.pi*f)))
```
