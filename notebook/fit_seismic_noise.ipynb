```python
import kontrol
import control
import numpy as np
import matplotlib.pyplot as plt


tf = kontrol.load_transfer_function("tf_xg_20240222.pkl")


s = control.tf("s")
wn = 0.159*2*np.pi
q = 10
dcgain = 2e-7
tf_fit = dcgain * wn**2 / (s**2 + wn/q*s + wn**2)

f = np.logspace(-3, 3, 1024)
plt.loglog(f, abs(tf(1j*2*np.pi*f)))
plt.loglog(f, abs(tf_fit(1j*2*np.pi*f)))
```
